```python
import numpy as np
import pandas as pd
import time
import requests
from bs4 import BeautifulSoup
```

# preparing the arrays

```python
product_name=[]
price_before_dsc=[]
dsc_percentage=[]
price_after_dsc=[]
product_review=[]
```

```python
page_num=0

while True:

    print(f'scrapping page num : {page_num}')
    
    url = f'https://www.jumia.com.eg/mlp-official-stores/?page={page_num}#catalog-listing'
    
    try:
        page = requests.get(url,timeout=10)
    except Exception as error:
        print(f'Error occured : {error}')
        
    soup = BeautifulSoup(page.text,'lxml')
    all_divs = soup.find_all('div',class_='info')

    if not all_divs:
        print('No more books')
        break
    
    for div in all_divs:
        condition_1=div.find('h3',class_='name')
        product_name.append(condition_1.text.strip()
                           if condition_1 else None)
    
        condition_2=div.find('div',class_='old')
        price_before_dsc.append(condition_2.text.strip()
                               if condition_2 else None)
    
        condition_3=div.find('div',class_='prc')
        price_after_dsc.append(condition_3.text.strip()
                              if condition_3 else None)
    
        condition_4=div.find('div',class_='bdg _dsct _sm')
        dsc_percentage.append(condition_4.text.strip()
                             if condition_4 else None)
    
        condition_5=div.find('div',class_='stars _s')
        product_review.append(condition_5.text.strip()
                             if condition_5 else None)
    page_num=page_num+1
    time.sleep(3)
```

# sving data in csv file

```python
data = {'prod_name':product_name,
        'prc_before_dsc':price_before_dsc,
        'prc_after_dsc':price_after_dsc,
        'dsc_percentage':dsc_percentage,
        'prod_review':product_review}
data_frame=pd.DataFrame(data)
data_frame.to_csv(r'C:\Users\Me\Downloads\Data_Internship\project_1\data_files\data.csv',index=False)
```
